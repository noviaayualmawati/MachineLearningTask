# 3. Data Preprocessing

3.1 Feature Engineering

```python
 #Pembuatan Kelompok Usia
df['age_group'] = pd.cut(df['usia'], bins=[0, 25, 35, 45, 55, 65, 100], labels=["0-25", "25-35", "35-45", "45-55", "55-65", "65+"])

# Encoding Gender
df['gender_encoded'] = pd.get_dummies(df['gender'], drop_first=True)

# Encoding Pendidikan
pendidikan_dummies = pd.get_dummies(df['pendidikan'], prefix='pendidikan')
df = pd.concat([df, pendidikan_dummies], axis=1)

# Encoding Status Nikah
status_nikah_dummies = pd.get_dummies(df['status_nikah'], prefix='status_nikah')
df = pd.concat([df, status_nikah_dummies], axis=1)

# Encoding Tipe Kartu Kredit
tipe_kartu_kredit_dummies = pd.get_dummies(df['tipe_kartu_kredit'], prefix='tipe_kartu_kredit')
df = pd.concat([df, tipe_kartu_kredit_dummies], axis=1)
```

```python
df['penghasilan_tahunan'] = df['penghasilan_tahunan'].replace('unknown', pd.NA) # Mengganti 'unknown' dengan NaN

df['penghasilan_tahunan'] = df['penghasilan_tahunan'].str.replace('$', '').str.replace(' ', '').str.replace('K', '000')# Normalisasi teks dalam kolom penghasilan_tahunan

# Mapping nilai kategori penghasilan tahunan ke nilai numerik
income_mapping = {
    'Lessthan40000': 40000,
    '40000-60000': 50000,
    '60000-80000': 70000,
    '80000-120000': 100000,
    '120000+': 120000
}

df['penghasilan_tahunan_num'] = df['penghasilan_tahunan'].map(income_mapping)

median_income = df['penghasilan_tahunan_num'].median()
df['penghasilan_tahunan_num'] = df['penghasilan_tahunan_num'].fillna(median_income)

# Menghitung pendapatan_per_tanggungan
df['pendapatan_per_tanggungan'] = df.apply(lambda row: row['penghasilan_tahunan_num'] / row['jumlah_tanggungan'] if row['jumlah_tanggungan'] > 0 else row['penghasilan_tahunan_num'], axis=1)

print(df[['penghasilan_tahunan', 'penghasilan_tahunan_num', 'pendapatan_per_tanggungan']].head())
```

```python
df['transaksi_per_produk'] = df['total_transaksi'] / df['jumlah_produk']
```

Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, RobustScaler, PowerTransformer

from sklearn.linear_model import LogisticRegression
```

```python
# Preprocessing pipeline for numerical features
num_preprocessor = Pipeline([("num_imputer",SimpleImputer(strategy='median')),
                             ("num_scaler",MinMaxScaler())])

cat_preprocessor = Pipeline([("cat_imputer",SimpleImputer(strategy='most_frequent')),
                             ("cat_encoder",OneHotEncoder())])

numeric_features = ['usia', 'lama_nasabah', 'jumlah_produk', 'bulan_nonactive', 'jumlah_kontak', 'total_limit_kredit', 'total_limit_kredit_dipakai', 'total_transaksi', 'jumlah_transaksi', 'rasio_pemakaian', 'pendapatan_per_tanggungan']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Preprocessing pipeline for categorical features
categorical_features = ['gender', 'pendidikan', 'status_nikah', 'tipe_kartu_kredit']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create the pipeline
#pipeline = Pipeline(steps=[
#    ('preprocessor', preprocessor),
#    ('classifier', LogisticRegression(random_state=42))
#])
complete_pipeline = Pipeline([("preprocess",preprocessor),
                              ("model",LogisticRegression())])
```
